```
# Q2 採樣出所有資料的性別、年齡、是否生存、Embarked
df_2_1 = pd.DataFrame(df.loc[:,('Sex','Age','Survived','Embarked')],
                   columns=['Sex','Age','Survived','Embarked'])
df_2_1
```

```
# Q2 找出性別的mode(眾數),mean(平均值)
df_2_2_Sex = pd.DataFrame(df.loc[:,'Sex'])

df_2_2_SexMode = df_2_2_Sex.mode()
print(df_2_2_SexMode)

df_2_2_SexMean = df_2_2_Sex.mean()
print(df_2_2_SexMean)

#df_2_2_Sex = pd.DataFrame({'lab':['male', 'female'], 'val':[10, 30]})
#df_2_2_Sex_Bar = df_2_2_Sex.plot.bar(x='Sex', y='count', rot=0)
#print(df_2_2_Sex_Bar)
```

```
# Q2 找出年齡的mode(眾數),mean(平均值)
df_2_2_Age = pd.DataFrame(df.loc[:,'Age'])

df_2_2_AgeMode = df_2_2_Age.mode()
print(df_2_2_AgeMode)

df_2_2_AgeMean = df_2_2_Age.mean()
print(df_2_2_AgeMean)
```

```
# Q2 找出是否生存的mode(眾數),mean(平均值)
df_2_2_Survived = pd.DataFrame(df.loc[:,'Survived'])

df_2_2_SurvivedMode = df_2_2_Survived.mode()
print(df_2_2_SurvivedMode)

df_2_2_SurvivedMean = df_2_2_Survived.mean()
print(df_2_2_SurvivedMean)
```

```
# Q2 找出Embarked的mode(眾數),mean(平均值)
df_2_2_Embarked = pd.DataFrame(df.loc[:,'Embarked'])

df_2_2_EmbarkedMode = df_2_2_Embarked.mode()
print(df_2_2_EmbarkedMode)

df_2_2_EmbarkedMean = df_2_2_Embarked.mean()
print(df_2_2_EmbarkedMean)
```
